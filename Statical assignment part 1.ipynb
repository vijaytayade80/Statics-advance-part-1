```python
Q3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution>
```

```python
from scipy.stats import bernoulli

def bernoulli_pdf(p):
    """
    Calculate the PDF of a Bernoulli distribution with given probability p.
    Returns a dictionary with outcomes {0: P(X=0), 1: P(X=1)}.
    """
    x = [0, 1]  # possible outcomes
    pdf = {val: bernoulli.pmf(val, p) for val in x}
    return pdf

# Example usage:
p = 0.7  # probability of success
pdf_result = bernoulli_pdf(p)
print(f"PDF of Bernoulli(p={p}):", pdf_result)
```

```python
Q6 Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution>
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the discrete uniform values (e.g., outcomes of a fair die: 1 to 6)
values = np.arange(1, 7)
# Each outcome has equal probability
probabilities = np.full_like(values, 1 / len(values), dtype=float)

# Calculate the cumulative probabilities (CDF)
cdf = np.cumsum(probabilities)

# Plot the CDF
plt.figure(figsize=(6, 4))
plt.step(values, cdf, where='mid', label='CDF', color='blue')
plt.title("CDF of Discrete Uniform Distribution (Fair Die)")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.ylim(0, 1.05)
plt.xticks(values)
plt.grid(True)
plt.legend()
plt.show()
```

```python
Q9 Write a Python function to calculate Z-scores from a dataset and plot them>
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores from a dataset and plot their distribution.
    
    Parameters:
        data (array-like): Input dataset
    Returns:
        z_scores (np.ndarray): Array of Z-scores
    """
    # Calculate Z-scores
    z_scores = zscore(data)

    # Plot the Z-scores
    plt.figure(figsize=(7, 4))
    sns.histplot(z_scores, bins=30, kde=True, color='orchid', edgecolor='black')
    plt.title("Z-Score Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.axvline(0, color='black', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=60, scale=15, size=1000)
z_scores = calculate_and_plot_zscores(sample_data)
```

```python
Q10 Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set parameters
np.random.seed(42)
population_size = 100000
sample_size = 30
num_samples = 1000

# Step 1: Generate a non-normal population (Exponential distribution)
population = np.random.exponential(scale=2, size=population_size)

# Step 2: Draw many samples and compute sample means
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# Step 3: Plot population vs. sample mean distribution
plt.figure(figsize=(14, 5))

# Left: Original non-normal distribution
plt.subplot(1, 2, 1)
sns.histplot(population, bins=50, kde=True, color='tomato')
plt.title("Original Population Distribution (Exponential)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)

# Right: Distribution of sample means
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=30, kde=True, color='seagreen')
plt.title(f"Sampling Distribution of the Mean (n={sample_size})")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
Q 14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution5
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    return (x - mean) / std

def plot_z_score(x, mean, std):
    z = calculate_z_score(x, mean, std)

    # Standard normal distribution curve
    x_values = np.linspace(-4, 4, 1000)
    y_values = norm.pdf(x_values, 0, 1)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, label='Standard Normal Curve', color='blue')
    plt.fill_between(x_values, y_values, where=(x_values < z), color='skyblue', alpha=0.6, label='Area Left of Z')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    
    plt.title(f'Z-score Visualization for X = {x}')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Z-score for X = {x} is: {z:.2f}")
    p_value = norm.cdf(z)
    print(f"Proportion of data less than X (P(Z < {z:.2f})) = {p_value:.4f}")
    print(f"Proportion of data greater than X (P(Z > {z:.2f})) = {1 - p_value:.4f}")

# Example usage
data_point = 75
population_mean = 70
population_std = 5

plot_z_score(data_point, population_mean, population_std)
```

```python
Q 15. Implement hypothesis testing using Z-statistics for a sample dataset5
```

```python
import numpy as np
from scipy.stats import norm

# Given values
population_mean = 100
population_std = 15
sample_size = 50
sample_mean = 104
alpha = 0.05

# Step 1: Calculate Z-statistic
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 2: Find critical Z for two-tailed test
z_critical = norm.ppf(1 - alpha / 2)

# Step 3: Decision
print(f"Z-statistic: {z_score:.4f}")
print(f"Critical Z-value (±): {z_critical:.4f}")

if abs(z_score) > z_critical:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")

# Optional: Calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))
print(f"P-value: {p_value:.4f}")
```

```python
Q 16. Create a confidence interval for a dataset using Python and interpret the result
```

```python
import numpy as np
from scipy import stats

# Sample dataset (e.g., test scores)
data = [68, 72, 75, 70, 74, 73, 69, 71, 77, 76]

# Step 1: Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
n = len(data)
confidence = 0.95

# Step 2: Calculate t-critical value
alpha = 1 - confidence
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Step 3: Margin of Error and Confidence Interval
margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 4: Output
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval ({confidence*100:.0f}%): ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
Q 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean5
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Step 1: Generate random data from a normal distribution
np.random.seed(42)  # for reproducibility
mu = 100     # true mean
sigma = 15   # true std deviation
n = 50       # sample size

data = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
confidence = 0.95
alpha = 1 - confidence
df = n - 1

# Step 3: t-critical value
t_critical = stats.t.ppf(1 - alpha/2, df)

# Step 4: Margin of error and confidence interval
margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 5: Output
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"{int(confidence*100)}% Confidence Interval for the Mean: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
Q 24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.
```

```python
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt

def analyze_normal_distribution(n=1000, mean=0, std=1, visualize=True):
    # Generate data from a normal distribution
    data = np.random.normal(loc=mean, scale=std, size=n)
    
    # Calculate skewness and kurtosis
    skewness = skew(data)
    kurt = kurtosis(data, fisher=False)        # Use fisher=False for normal kurtosis (≈ 3)
    excess_kurtosis = kurtosis(data, fisher=True)  # Use fisher=True for excess kurtosis (≈ 0)

    # Print results
    print(f"Skewness: {skewness:.4f} (≈ 0 for normal)")
    print(f"Kurtosis: {kurt:.4f} (≈ 3 for normal)")
    print(f"Excess Kurtosis: {excess_kurtosis:.4f} (≈ 0 for normal)")

    # Optional: Visualize the distribution
    if visualize:
        plt.figure(figsize=(8, 4))
        plt.hist(data, bins=40, color='skyblue', edgecolor='black', density=True)
        plt.title('Histogram of Normally Distributed Data')
        plt.xlabel('Value')
        plt.ylabel('Density')
        plt.grid(True)
        plt.show()

# Call the function
analyze_normal_distribution()
```
